# Stock symbols

```python
start = time.time()

alpha = list(string.ascii_uppercase)
digit = ['0','3','4','5','6','7','8','9']
combined = alpha + digit

symbols = []

# Loop through to get the stock symbols on each page
# from the table and store them in a list
for each in combined:
    url = 'http://eoddata.com/stocklist/LSE/{}.html'.format(each)
    resp = requests.get(url)
    site = resp.content
    soup = BeautifulSoup(site, 'html.parser')
    table = soup.find('table', {'class': 'quotes'})
    for row in table.findAll('tr')[1:]:
        symbols.append(row.findAll('td')[0].text.rstrip())       


# Remove the extra letters on the end
symbols_clean = []

for each in symbols:
    each = each.replace('.', '-')
    symbols_clean.append((each.split('-')[0]))
    
end = time.time()

time_to_run = end - start
time_to_run
print("There are " + str(len(symbols_clean)) + " symbols")
```

# Getting quotes

```python
#api_key = 'pk_02909ca51d914e28a8d77250aaae6a56'
sandbox_key = 'Tpk_5e1f553159ea4051b2c2c6ee92273bc3'

def quotes_request(symbols):
    
    quotes = []
    valid = 0
    invalid = 0
    
    for symbol in symbols:
        url = 'https://sandbox.iexapis.com/stable/stock/{}/quote?token={}'.format(symbol, sandbox_key)
    
        request = requests.get(url)
    
        if request.status_code == 404:
            invalid += 1 # Track number of invalid responses
            continue # Skip to next symbol if HTTP response is 404 Not Found
        
        valid += 1  # Track number of valid responses
        quotes.append(request.json())
        
    print("There are " + str(valid) + " valid symbols and " + str(invalid) + " invalid symbols")

    return pd.DataFrame(quotes)


df = quotes_request(symbols[0:100]) 
    
df   
```
