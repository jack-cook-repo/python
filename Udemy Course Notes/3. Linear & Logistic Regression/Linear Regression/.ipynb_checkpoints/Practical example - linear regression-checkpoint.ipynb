### 5. No multicollinearity

```python
data_cleaned.columns.values
```

Year & mileage are likely to be correlated. Sklearn doesn't have built in checks for multicollinearity, instead we can use statsmodels - for the VIF (Variance Inflation Factor).

*The VIF may be calculated for each predictor by doing a linear regression of that predictor on all the other predictors, and then obtaining the R2 from that regression. The VIF is just 1/(1-R2).*

*Its called the variance inflation factor because it estimates how much the variance of a coefficient is inflated because of linear dependence with other predictors. Thus, a VIF of 1.8 tells us that the variance (the square of the standard error) of a particular coefficient is 80% larger than it would be if that predictor was completely uncorrelated with all the other predictors.*

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

#We can only check continuous variables
variables = data_cleaned[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['features'] = variables.columns
```

```python
vif
```

VIF ranges from 1 to infinity, so 1 is the lowest value.

Values between 1 and ~10 are ok for this case

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
```

# Creating dummy variables

```python
data_no_multicollinearity.Brand.unique()
```

We only need to create **6** dummy variables, because if all other 6 brand dummy values are 0 - we can determine that it is the 7th remaining brand.

Also, if we had 7 dummy variables, the 7th would have perfect collinearity as it could be predicted entirely by the 6 other dummy variables alone.

```python
data_cleaned_with_dummies = pd.get_dummies(
    data_no_multicollinearity, 
    drop_first=True) #This returns k-1 columns from k categorical variables
```

```python
data_cleaned_with_dummies.head()
```

```python
#We can now check all variables VIF
variables2 = data_cleaned_with_dummies.loc[: ,data_cleaned_with_dummies.columns != 'log_price']
vif2 = pd.DataFrame()
vif2['VIF'] = [variance_inflation_factor(variables2.values, i) for i in range(variables2.shape[1])]
vif2['features'] = variables2.columns
vif2
```

# Regression model!

### Declare inputs and targets

```python
targets = data_preprocessed.log_price
inputs = data_preprocessed.drop(['log_price'], axis=1)
```

### Scale data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

### Train/test split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets,
                                                   test_size = 0.2,
                                                   random_state = 47)
```

### Create regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y train)', size = 20)
plt.ylabel('Predictions (y hat)', size = 20)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

### Look at residuals

```python
sns.distplot(y_train - y_hat)
plt.title('Residuals PDF', size = 10)
```

We can see that our errors are **normally distributed**, meeting the requirement for normality (requirement 3 of regression).

But we can see that there is a larger negative tail - so our model is likely to overestimate price rather than underestimate.

```python
reg.score(x_train, y_train)
```

### Find weights and bias

```python
reg.intercept_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

**For a continuous variable:** Negative weight means as a feature increases, the price of the car decreases

**For a discrete (dummy) variable:** Negative weight shows a respective category is more cheaper than the omitted category (eg. Mitsubishi is cheaper than Audi). For Mercedes/BMW the weights are positive, but *only just*, so we can't definitively say that they are more expensive

Remember, dummy weights can only be compared to the other dummys in that category group.

For continuous variable - you can see that mileage has by far the biggest impact

## Testing 

```python
y_hat_test = reg.predict(x_test)
plt.scatter(y_test, y_hat_test, alpha=0.4)
plt.xlabel('Targets (y_test)', size=20)
plt.ylabel('Predictions (y_hat)', size=20)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_performance = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_performance.head()
```

```python
#If we don't drop the indices, then the join below won't work!
y_test = y_test.reset_index(drop=True)

df_performance['Target'] = np.exp(y_test)
df_performance.head()
```

```python
df_performance['Residual'] = df_performance.apply(lambda row: row['Target'] - row['Prediction'], axis=1)
df_performance['Residual_pct'] = df_performance.apply(lambda row: row['Residual'] / row['Target'] * 100, axis=1)
df_performance.head()
```
