```python
# Open file and parse contents
file = open('day7_inputs.txt')
content_raw = [line.strip() for line in file]
#content_raw
```

Part 1

```python
import re
content_split = []

# Split parent / children
for c in content_raw:
    parent, children = c.split(' bags contain ')
    
    # Split children into sub list
    child_list = children.split(',')
    child_list = [ch.replace('.','').replace('bags','').replace('bag','').strip() for ch in child_list]
    child_list = [re.search('[a-z]+\s[a-z]+', ch).group() for ch in child_list]
    
    content_split.append([parent, child_list])
```

```python
# Create lookup dict
dict_content = dict(content_split)
dict_content
```

```python
# Get rows that either have no children, or already have gold bag as one of their children
colours_w_no_children = [val[0] for val in dict_content.items() if 'no other' in val[1]]
colours_w_shiny_gold = [val[0] for val in dict_content.items() if 'shiny gold' in val[1]]
#colours_w_shiny_gold.append('shiny gold')

keys_to_loop = list(dict_content.keys())

# Don't do unnecessary loops
keys_to_loop = [k for k in keys_to_loop if k not in colours_w_no_children and k not in colours_w_shiny_gold]

# Setup for recursive loop
keys_to_loop_next = keys_to_loop[:]
parent_child_dict = dict_content.copy()
new_parent_child_dict = {}

for i in range(30):
    if len(keys_to_loop) == 0:
        print('Loop finished on number {}'.format(i+1))
        break
        
    for k in keys_to_loop:
        next_colour_list = []
        children = parent_child_dict[k]

        # Get all child colours of children
        for c in children:
            if c != 'no other':
                next_colour_list.extend(dict_content[c])

        next_colour_list = list(set(next_colour_list))
        
        # If there are no more child colours to pursue, remove colour from loop
        if len(next_colour_list) == 0:
            keys_to_loop_next.remove(k)
        # If shiny gold appears, remove colour from next loop and add to list of approved colours
        elif 'shiny gold' in next_colour_list:
            keys_to_loop_next.remove(k)
            colours_w_shiny_gold.extend([k])
        else:
            # Append to next dictionary if the loop is to continue
            new_parent_child_dict[k] = next_colour_list
    
    # restart process
    keys_to_loop = keys_to_loop_next[:]
    parent_child_dict = new_parent_child_dict.copy()
    new_parent_child_dict = {}
   # if len(colours_w_shiny_gold) = len(new_colours_with_shiny_gold)
    
print(len(colours_w_shiny_gold))
```

Part 2

```python
import re
content_split = []

# Split parent / children
for c in content_raw:
    parent, children = c.split(' bags contain ')
    
    # Split children into sub list
    child_list = children.split(',')
    child_list = [ch.replace('.','').replace('bags','').replace('bag','').strip() for ch in child_list]
    child_colours = [re.search('[a-z]+\s[a-z]+', ch).group() for ch in child_list]
    child_numbers_prelim = [re.search('[0-9]+', ch) for ch in child_list]
    child_numbers = [0 if ch is None else ch.group() for ch in child_numbers_prelim]
    child_colours_numbers = list(zip(child_colours, child_numbers))
    
    content_split.append([parent, child_colours_numbers])

content_split[:5]
```

```python
dict_content = dict(content_split)
```

```python
keys_to_loop = ['shiny gold']
total_bags = 0

# Setup for recursive loop
keys_to_loop_next = keys_to_loop[:]
parent_child_dict = dict_content.copy()
new_parent_child_dict = {}

for i in range(30):
    if len(keys_to_loop) == 0:
        print('Loop finished on number {}'.format(i+1))
        break

    for k in keys_to_loop:
        next_colour_num_list = []
        children = parent_child_dict[k]

        # Get all child colours of children and number of bags
        for c in children:
            colour = c[0]
            num = int(c[1])
            print(colour, num)
            total_bags += num

            if colour != 'no other':
                # Get numbers and colours
                ch_nums = [int(col[1]) for col in dict_content[colour]]
                ch_colours = [col[0] for col in dict_content[colour]]
                ch_nums = [n * num for n in ch_nums]
                next_colour_num_list.extend(list(zip(ch_colours, ch_nums)))

        next_colour_num_list = list(set(next_colour_num_list))
        
        # If there are no more child colours to pursue, remove colour from loop
        if len(next_colour_num_list) == 0:
            keys_to_loop_next.remove(k)
        else:
            # Append to next dictionary if the loop is to continue
            new_parent_child_dict[k] = next_colour_num_list
    
    # restart process
    keys_to_loop = keys_to_loop_next[:]
    parent_child_dict = new_parent_child_dict.copy()
    new_parent_child_dict = {}

    
print(total_bags)
```
