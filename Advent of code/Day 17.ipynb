```python
with open('day17_inputs.txt') as f:
    content = [l.strip('\n') for l in f]

for l in content:
    print(l)
```

#### Part 1 

```python
neighbour_relative_co_ords = (-1,0,1)
neighbour_rel_co_ord_list = []

import itertools

for co_ord_combo in itertools.product(neighbour_relative_co_ords,
                                      neighbour_relative_co_ords,
                                      neighbour_relative_co_ords):
    if co_ord_combo != (0, 0, 0):
        neighbour_rel_co_ord_list.append(list(co_ord_combo))
        print(list(co_ord_combo))
```

```python
def get_cube_co_ords(cube_boundaries):
    # Turns x,y,z boundaries into cube co-ordinates as a list
    iterable = itertools.product(list(range(cube_boundaries[0])),
                                 list(range(cube_boundaries[1])),
                                 list(range(cube_boundaries[2])))
    
    return [list(coord) for coord in iterable]
    

cube_width = len(content[0])
cube_height = len(content)

cube_boundaries = [cube_width, # width (x)
                   cube_height, # height (y)
                   1] # depth (z)

cube_coords = get_cube_co_ords(cube_boundaries)
                   
import copy

# Initialise 'cube' as 2d plane
cube = np.array(content)

# Loop through cycles
for i in range(6):
    
    activated_cells = 0
    
    # Get new boundaries and update seed cube's co-ordinates
    new_cube_boundaries = [b+2 for b in cube_boundaries]
    new_cube_max_x, new_cube_max_y, new_cube_max_z = new_cube_boundaries
    new_cube_coords = get_cube_co_ords(new_cube_boundaries)
    seed_cube_coords = [[coord[0]+1,coord[1]+1,coord[2]+1] for coord in cube_coords]

    # Set new cube as as dots
    new_cube = np.full((new_cube_max_z,new_cube_max_y,new_cube_max_x),'.')

    # Update cube dots with seed cube values where they overlap
    for coord in new_cube_coords:
        x,y,z = coord
        if coord in seed_cube_coords:
            # Handle fact we start with 2d array on first loop
            if i == 0:
                new_cube[z][y][x] = cube[y-1][x-1]
            else:
                new_cube[z][y][x] = cube[z-1][y-1][x-1]

    # Set up cube to be altered
    new_cube_updated = new_cube.copy()

    # Loop through new cube again, this time recording neighbours activates
    for coord in new_cube_coords:
        x,y,z = coord
        neighbours_activated = 0

        for neighbour in neighbour_rel_co_ord_list:
            x_diff, y_diff, z_diff = neighbour
            new_x = x+x_diff
            new_y = y+y_diff
            new_z = z+z_diff
            min_coord = min(new_x, new_y, new_z)

            # Check if neighbour co-ordinate is out of bounds
            if min_coord < 0 or new_x > new_cube_max_x-1 or new_y > new_cube_max_y-1 or new_z > new_cube_max_z-1:
                continue
            # Check if a neighbouring point is activated
            elif new_cube[new_z][new_y][new_x] == '#':
                neighbours_activated+=1

        # Apply update rules
        if new_cube[z][y][x] == '#' and neighbours_activated not in [2,3]:
            new_cube_updated[z][y][x] = '.'

        if new_cube[z][y][x] == '.' and neighbours_activated == 3:
            new_cube_updated[z][y][x] = '#'
        
        # Count activated cells
        if new_cube_updated[z][y][x] == '#':
            activated_cells+=1
    
    
    # Set up variables for next loop
    cube_boundaries = copy.deepcopy(new_cube_boundaries)
    cube_coords = copy.deepcopy(new_cube_coords)
    cube = copy.deepcopy(new_cube_updated)

    print('loop {} complete, cells active {}'.format(i+1, activated_cells))
```
